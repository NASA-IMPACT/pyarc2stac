```python
from pyarc2stac.ArcReader import ArcReader
```

```python
image_server_url = "https://gis.earthdata.nasa.gov/image/rest/services/POWER/POWER_901_MONTHLY_RADIATION_UTC/ImageServer"
```

```python
reader = ArcReader(
    server_url = image_server_url
)
```

```python
collection = reader.generate_stac()
```
