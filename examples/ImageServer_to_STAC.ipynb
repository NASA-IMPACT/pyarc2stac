```python
from pyarc2stac.ArcReader import ArcReader
```

```python
image_server_url = 'https://maps.disasters.nasa.gov/ags03/rest/services/NRT/lis_vsm_percentile_100cm/ImageServer'
```

```python
reader = ArcReader(
    server_url = image_server_url
)
```

```python
collection = reader.generate_stac()
```

```python
collection.to_dict()
```

```python
reader.save_collection_to_json()
```
